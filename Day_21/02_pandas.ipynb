# Task 1
Retrieve and list the entries, sorting them from highest to lowest by `tip_percent`, excluding the first five results, specifically for weekends (Saturday and Sunday).

```python

ordered_df = tips_df.sort_values(by = "tip_percentage", ascending = False)
ordered_df[(ordered_df["day"] == "Sat")|(ordered_df["day"] == "Sun")].iloc[5:]
```

## Task 2
 
Retrieve entries for Thursday lunch with party sizes greater than 3, sort them by tip (descending), Retrieve only 2nd & 3rd rank

```python

#o_df = thursday_lunch_df.sort_values(by = "tip", ascending = False)

ordered_df = tips_df.sort_values(by="tip", ascending=False)

thursday_lunch_df = ordered_df[
    (ordered_df["day"] == "Thur") & (ordered_df["time"] == "Lunch") & (ordered_df["size"] > 3)
]

thursday_lunch_df.iloc[1:3]
```

```python
# o_df = thursday_lunch_df.sort_values(by = "tip", ascending = False)

ordered_df = tips_df.sort_values(by="tip", ascending=False)
thursday_lunch_df = ordered_df[
    ordered_df["day"] == "Thur")
    & (ordered_df["time"] == "Lunch")
    & (ordered_df["size"] > 3)
]

thursday_lunch_df.iloc[1:3]
```

## Task
 
Draw a bar graph for Average total Bill for each Day

```python
y_axis = tips_df.groupby("day")["total_bill"].mean()
x_axis = y_axis.index  # to start with index 1
# x_axis = np.arange(5, 10)

plt.bar(x_axis, y_axis, color="green", alpha=0.4) # alpha value makes colour lighter
plt.xlabel("Day")
plt.ylabel("Average Bill")
plt.title("Average Total Bill For Every Day")
```

```python
y_axis = tips_df.groupby("day")["total_bill"].mean()
y_axis.plot(kind="bar", xlabel = "Days", ylabel="Average Sale", title= "Average Sales in Week", color="orange", alpha=0.6)
```

```python
y_axis = tips_df.groupby("day")["total_bill"].mean()
y_axis.plot(
    kind="pie",
    autopct="%1.1f%%",
    title="Average Sales in Week",
)
```
